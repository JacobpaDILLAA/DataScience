```python
from nltk.stem.porter import PorterStemmer # Stem operation
import os
import glob
from pathlib import Path

inverted_list = []
index = []


def main():
    global inverted_list
    pathen = './data'
    for filename in glob.glob(os.path.join(pathen, '*.txt')):
        with open(filename, 'r') as f:
            text = f.read()
            word = stem_word(text)
            inverted_list.append(word)

            
def stem_word(word):
    porter = PorterStemmer()
    stemmed = porter.stem(word)
    return stemmed


# This method is to read to query file to set the operation
def read_query():
    global index
    for count, query in enumerate(open('./query.txt', 'r')):
        for term in query.split():
            index.append(term)
        term1 = index[0]
        operation = index[1]
        term2 = index[2]
        

def writefile(inverted_answer):
    temp = open("./answer.txt", "a+")
    temp.write(f"{inverted_answer}\n")



main()
read_query()



p = './index_Padil'
wr = open(p,"a+",encoding="utf8")
for term in inverted_list:
    temp = str(term)
    wr.write(temp)


# This method is here to aid in the 'or' operation in the given query.
def or_operation(term1, term2):
    container = []
    if term1 is None or term2 is None:
        return -1

    
    stem1 = stem_word(term1)
    stem2 = stem_word(term2)
    container.append(stem1 + sorted(set(stem2) - set(stem1)))
    sorted_container = sorted(container)
    return sorted_container

# This method is here to aid in the 'and' operation in the given query.
def and_operation(term1, term2):
    container = []
    stem1 = stem_word(term1)
    stem2 = stem_word(term2)

    for i in range(stem1):
        for j in range(stem2):
            if i < len(stem1) and j < len(stem2):
                if stem1[i] == stem2[j]:
                    container.append(stem1[i])
                    i += 1; j += 1
                elif stem1[i] < stem2[j]:
                    i += 1
                else:
                    j += 1

    return container
```
