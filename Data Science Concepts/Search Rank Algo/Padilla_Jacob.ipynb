```python
import pandas as pd
from os.path import join
import requests
from bs4 import BeautifulSoup
import html2text
from urllib.request import urlopen

#relative path 
index_path = "./indexedFiles/"

req = requests.get("https://www.utep.edu/cs/people/index.html")
soup = BeautifulSoup(req.text, 'lxml') # HTML text and lxml parser

#Traversing the HTML Tree
divs = soup.findAll("div", {"class": "col-md-6"}) #Grabs info of faculty

#used to store information from webpage
raw1=[]
raw2=[]
raw3=[]
raw4=[]
raw5=[]
raw6=[]

for div in divs: 
    #Name
    facultyName = div.find("h3", {"class":"name"})
    
    #Office
    facultyAddress = div.find("span", {"class":"address"})
    
    #Email
    facultyEmail = div.find("span", {"class":"email"})
    
    #Phone
    facultyPhone = div.find("span", {"class":"phone"})
    
    #Email variable for extracting user name
    UserEmail = div.find("span", {"class":"email"})
      
    #Title
    facultyTitle = div.find("span", {"class": "Title"})
    if facultyTitle is not None:
        if "Professor" in facultyTitle.text:
            
            raw2.append(facultyTitle.text)
        
            if facultyName is not None:
                raw1.append(facultyName.text)
    
            if facultyAddress is not None:
                raw3.append(facultyAddress.text)
    
            if facultyEmail is not None:
                raw4.append(facultyEmail.text)
        
            if facultyPhone is not None:
                #res = [re.findall(r'(\w+?)(\d+)', facultyPhone.text)[0] ]
                raw5.append(facultyPhone.text)
        
    
            #extract username from email
            temp = UserEmail.text.split("@")[0]
            name = temp
            
            # Web Page
            facultyWeb = div.findAll('a')
            facultyURL = facultyWeb[len(facultyWeb) - 1].get("href")
            #if len(facultyURL) > 0 :
            if len(facultyURL) is not None:
                raw6.append(facultyURL)
                if len(facultyURL) > 0:
                    WebPageContent = requests.get(facultyURL) #html format
                    content = BeautifulSoup(WebPageContent.text, "lxml") #plain text
                    name = temp+".txt" #name of professor
                    file = open(join(index_path,name),'w') #open professor folder on Desktop
                    n = file.write(content.text)
                    file.close()
                
#contains information for the dataframe
raw_data = {'Name':raw1[0:], 'Title':raw2[0:], 'Office':raw3[0:], 
            'Email':raw4[0:], 'Phone':raw5[0:], 'website':raw6[0:]}

#send data to dataframe
professor_dataframe = pd.DataFrame(raw_data)

#pickle dataframe
professor_dataframe.to_pickle("./professors.pkl")

#testing to see if functioning correctly
professor_dataframe

#used to read the pickled dataframe
#unpickled_professors = pd.read_pickle("./professors.pkl")
#unpickled_professors
```

```python
from matchup.structure.vocabulary import Vocabulary
from matchup.structure.query import Query
from matchup.models.algorithms import Boolean
from matchup.models.algorithms import Vector
from matchup.models.algorithms import Probabilistic

import time
import pandas as p

vocabulary = Vocabulary('results',stopwords='stopWords/stopwords.txt') 
vocabulary.import_folder('indexedFiles')
vocabulary.index_files()
vocabulary.save()
vocabulary.import_collection()

fileName = 'queries.txt'
fileObj = open(fileName, "r") #opens the file in read mode
words = fileObj.read().splitlines() #puts the file into an array
fileObj.close()

f_name = './results/res.txt'
data_file = []   

################################ Boolean Implementation ############################

print(" ----- Boolean Model ----- ")
Bstart = time.time()

#process queries.
query = Query(vocabulary=vocabulary)

for t in words:
    answer = t
    query.ask(answer)
    results = query.search(model=Vector())
    data_file.append(results)
    print()
    print("query term:",answer, " ")
    print(" ",results, " ")
    print()
    
Bend = time.time()
Btemp = Bend - Bstart
Btotal = round(Btemp, 3)
print("----- Time for Boolean Model ",Btotal," -----")
print()

################################ Vector Implementation ############################

print(" ----- Vector Space Model ----- ")
Vstart = time.time()

#process queries.
query = Query(vocabulary=vocabulary)

for t in words:
    answer = t
    query.ask(answer)
    results = query.search(model=Vector())
    data_file.append(results)
    print()
    print("query term:",answer, " ")
    print(" ",results, " ")
    print()
    
Vend = time.time()
Vtemp = Vend - Vstart
Vtotal = round(Vtemp, 3)
print("----- Time for Vector Model ",Vtotal," -----")
print()

################################ Probalistic Implementation ############################

print(" ----- Probalistic Model ----- ")
Pstart = time.time()

#process queries.
query = Query(vocabulary=vocabulary)

for t in words:
    answer = t
    query.ask(answer)
    results = query.search(model=Probabilistic())
    data_file.append(results)
    print()
    print("query term:",answer, " ")
    print(" ",results, " ")
    print()
    
Pend = time.time()
Ptemp = Pend - Pstart
Ptotal = round(Ptemp, 3)
print("----- Time for Probalistic Model ",Ptotal,"-----")
print()


with open(f_name, 'w') as f:
    for item in data_file:
        f.write("%s\n" % item)
        
print()
print("Overall Times")
print("Boolean Time:", Btotal)
print("Vector Time:", Vtotal)
print("Probalistic Time:", Ptotal)



```
